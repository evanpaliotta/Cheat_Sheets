```python
Vocab:
1. Print: to print either a "word" or variable
2. Comment: in-code documentation, or can be used to prevent Python from executing the code. Multi line comment requires 3 sets of quotes.
3. Variable: containers for storing data values.  Can store one variable or multiple variables. To combine both text and a variable use print ("this is the text" + variable)
4. Plus_Sign: either a mathematical operator or a combiner of variables...concatinate
5. Types: text type (str), numeric_types (int, float, complex), sequence_types (list, tuple, range), mapping type (dict), set_types (set, frozenset), Boolean_types (bool), binary_types (byte, bytearray, memoryview).....if you want to set the data type you can use the datatype (specify) for ex. x = list(("apple", "banana", "cherry"))
6. Comparison_Operators: compares one value to another..... ==: equal to, !=: not equal to, the rest of the classic comparisons like <=, >=, <, >
7. Assignment_Operators: assigns a variable to something .....=...... the rest of the assignment operators stem from =...ex. x += 3 means x = x + 3
8. Logical_Operators: used to combine conditional statements.....'and' returns true if both statements are true, 'or' returns true if either statement is true, 'not' reverses the result -- returns False is statement is True
9. Identity_Operators: used to compare the objects, not if they are equal, but if they are actually the same object.....'is' returns True if both variables are the same object, 'is not' returns true if both variables are not the same object
10. List: collection which is ordered and mutable
11. Tuple: collection which is ordered and immutable
12. Set: collection which is unordered and unindexed
13. Dictionary: collection which is unordered, changeable, and indexed...written in {curly brackets} and have keys and values
14. Array_Functions: an array is either a list, tuple, set, or dictionary..... .append adds an item to an array.... .insert inserts an item in a specific spot.... .remove removes an item and you do not need to specify where it is in the array..... .pop removes a specific indexed item
15. If...Else...Elif: if initiates a comparison or a filtering process....else keyword catches anything which isn't caught by the previous conditions.... elif is Python's way of saying "if the previous conditions were not true, then try this condition"
16. While_Loops: we can execute a set of statements as long as a condition is true
17. For_Loops: used for iterating over a sequence... we can execute a set of statements, one for each item in the sequence
18. Function: a function is a block of code which only runs when it is called.... you can pass data, known as parameters, into a function.... a function can return data as a result.... a function is defined using the 'def' keyword...to call the function, use the functions name followed by () 
19. Chained Conditionals: multiple conditions in one if statement...and, or, not
```
